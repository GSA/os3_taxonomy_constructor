### This should pull Feature Bullets from product page and give a cleanish text for a single ASIN. Amazon started doing robot tests and broke it. Currently this is using ArmMount array. Next step would be adjust to to use asin_array once that is completed. 

```python
feature_array=[] #An array to store details of a single product.
asin_map = {k:None for k in ArmMount}
for asin in ArmMount:
    amazon_url="https://www.amazon.com/dp/"+asin #The general structure of a url
    response = requests.get(amazon_url, headers=headers) #get the response
    #description_array.append((soup.find(id="productDescription")).p.contents[0].strip()) #Extracting the details
    soup = BeautifulSoup(response.content, "lxml")
    feature_temp = (soup.find(id="feature-bullets"))
    try:
        feature_temp = feature_temp.text.replace('\r', '').replace('\n', '').replace('\t', '')
        feature_temp = re.sub(' +',' ',feature_temp)
        feature_temp = feature_temp[feature_temp.find("})"):]
        print(feature_temp)
    except AttributeError:
        pass
    asin_map[asin] = feature_temp
    time.sleep(10)
```
